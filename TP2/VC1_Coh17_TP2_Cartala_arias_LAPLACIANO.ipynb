## Parte 1: Implementar el algoritmo de enfoque que ya usamos en los experimentos (frame completo, roi y matrices de enfoque) pero con otra métrica (apéndice de: 'Analysis of focus measure operators in shapefrom focus').
Implementamos en este caso la métrica de enfoque usando la 'Varianza del Laplaciano'.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Leer el video
video_path = 'focus_video.mov'
cap = cv.VideoCapture(video_path)

focus_scores = [] # Lista para almacenar los valores de la métrica de enfoque frame a frame
frame_count = 0 # Contador de frames

# Obtener el ancho, alto y fps del video original
frame_width = int(cap.get(cv.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv.CAP_PROP_FPS)

# Definir el códec y crear el objeto VideoWriter para el nuevo video
output_path = 'focus_video_output_exp1_LAPLACIAN.avi'
fourcc = cv.VideoWriter_fourcc(*'XVID')
out = cv.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

while cap.isOpened():
    ret, frame = cap.read() # Leer un frame del video
    if not ret: # Si no se pudo leer el frame, terminar el loop
        break
    
    # Convertir el frame a escala de grises
    gray_frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    
    
    
    # Calcular la varianza del Laplaciano en el frame
    laplacian_var = cv.Laplacian(gray_frame, cv.CV_64F).var()

    focus_scores.append(laplacian_var) # Guardar el valor de la métrica

    frame_count += 1 # Incrementar el contador de frames

max_focus_score = max(focus_scores)
min_focus_score = min(focus_scores)
print(f'El valor máximo de la métrica de enfoque VARIANZA DEL LAPLACIANO es: {max_focus_score:.2f}')
print(f'El valor mínimo de la métrica de enfoque VARIANZA DEL LAPLACIANO es: {min_focus_score:.2f}')

# Graficar la evolución de la métrica frame a frame
plt.plot(focus_scores)
plt.title('Evolución de la métrica de enfoque frame a frame')
plt.xlabel('Frame')
plt.ylabel('Métrica de Enfoque')
plt.show()

# Volver a abrir el video para añadir la superposición de las cajas y el texto
cap = cv.VideoCapture(video_path)
frame_count = 0  # Reiniciar el contador de frames

while cap.isOpened():
    ret, frame = cap.read()  # Leer un frame del video
    if not ret:
        break

    # Obtener el valor de la métrica de enfoque para el frame actual
    focus_measure = focus_scores[frame_count]

    # Determinar si la métrica es mayor o menor a 80 (valor elegido para considerar un frame desenfocado)
    if focus_measure < 80:
        # Dibujar un bounding box de color rojo que ocupe todo el ancho y alto del frame
        cv.rectangle(frame, (0, 0), (frame_width, frame_height), (0, 0, 255), 10)
        
        # Agregar el texto 'DESENFOCADO' con el valor de la métrica centrado en el frame
        text1 = f'DESENFOCADO'
        text2 = f'Metrica: {focus_measure:.2f}'
        font = cv.FONT_HERSHEY_SIMPLEX
        text_size1 = cv.getTextSize(text1, font, 2, 3)[0]
        text_size2 = cv.getTextSize(text2, font, 1, 3)[0]
        text1_x = (frame_width // 2) - (text_size1[0] // 2)
        text1_y = (frame_height // 4) - text_size1[1] // 2
        text2_x = (frame_width // 2) - (text_size2[0] // 2)
        text2_y = (frame_height // 4) * 3 - text_size2[1] // 2
        cv.putText(frame, text1, (text1_x, text1_y), font, 2, (0, 0, 255), 3, cv.LINE_AA)
        cv.putText(frame, text2, (text2_x, text2_y), font, 1, (0, 0, 255), 3, cv.LINE_AA)
    else:
        # Dibujar un bounding box de color verde que ocupe todo el ancho y alto del frame
        cv.rectangle(frame, (0, 0), (frame_width, frame_height), (0, 255, 0), 10)
        
        # Agregar el texto 'ENFOCADO' con el valor de la métrica centrado en el frame
        text1 = f'ENFOCADO'
        text2 = f'Metrica: {focus_measure:.2f}'
        font = cv.FONT_HERSHEY_SIMPLEX
        text_size1 = cv.getTextSize(text1, font, 2, 3)[0]
        text_size2 = cv.getTextSize(text2, font, 1, 3)[0]
        text1_x = (frame_width // 2) - (text_size1[0] // 2)
        text1_y = (frame_height // 4) - text_size1[1] // 2
        text2_x = (frame_width // 2) - (text_size2[0] // 2)
        text2_y = (frame_height // 4) * 3 - text_size2[1] // 2
        cv.putText(frame, text1, (text1_x, text1_y), font, 2, (0, 255, 0), 3, cv.LINE_AA)
        cv.putText(frame, text2, (text2_x, text2_y), font, 1, (0, 255, 0), 3, cv.LINE_AA)

    # Escribir el frame en el nuevo video
    out.write(frame)
    frame_count += 1

# Liberar los objetos VideoCapture y VideoWriter
cap.release()
out.release()
```

### Parte 1b: Medición sobre una ROI en el Centro del Frame
Calculamos la métrica solo en una región de interés (ROI) en el centro del frame. El área de la ROI será del 10% del área total del frame.

```python
cap = cv.VideoCapture(video_path)
focus_scores_roi = []

# Definir el códec y crear el objeto VideoWriter para el nuevo video
output_path = 'focus_video_output_exp2_roi_LAPLACIANO.avi'
fourcc = cv.VideoWriter_fourcc(*'XVID')
out = cv.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

while cap.isOpened():
    ret, frame = cap.read() # Leer un frame del video
    if not ret:
        break
    
    # Convertir el frame a escala de grises
    gray_frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    
    # Definir ROI en el centro del frame (10% del área total)
    h, w = gray_frame.shape # Alto y ancho del frame
    roi_h, roi_w = int(0.1 * h), int(0.1 * w) # Alto y ancho de la ROI
    start_x, start_y = (w - roi_w) // 2, (h - roi_h) // 2 # Coordenadas de inicio de la ROI
    roi = gray_frame[start_y:start_y + roi_h, start_x:start_x + roi_w] # Extraer la ROI
    
    # Calcular la varianza del Laplaciano en el frame
    laplacian_var = cv.Laplacian(gray_frame, cv.CV_64F).var()
    focus_scores_roi.append(laplacian_var) # Guardar el valor de la métrica

max_focus_score_roi = max(focus_scores_roi)
min_focus_score_roi = min(focus_scores_roi)
print(f'El valor máximo de la métrica de enfoque VARIANZA DEL LAPLACIANO para la ROI es: {max_focus_score_roi:.2f}')
print(f'El valor mínimo de la métrica de enfoque VARIANZA DEL LAPLACIANO para la ROI es: {min_focus_score_roi:.2f}')

# Graficar la evolución de la métrica en la ROI
plt.plot(focus_scores_roi)
plt.title('Evolución de la métrica de enfoque sobre ROI en el centro')
plt.xlabel('Frame')
plt.ylabel('Métrica de Enfoque ROI')
plt.show()

# Volver a abrir el video para añadir la superposición de las cajas y el texto
cap = cv.VideoCapture(video_path)
frame_count = 0  # Reiniciar el contador de frames

while cap.isOpened():
    ret, frame = cap.read()  # Leer un frame del video
    if not ret:
        break

    # Obtener el valor de la métrica de enfoque para el frame actual
    focus_measure = focus_scores_roi[frame_count]

    # Determinar si la métrica es mayor o menor a 450000000 (valor elegido para considerar un frame desenfocado)
    if focus_measure < 80:
        # Dibujar un bounding box de color rojo que ocupe todo el ancho y alto del frame
        cv.rectangle(frame, (start_x, start_y), (start_x + roi_w, start_y + roi_h), (0, 0, 255), 1)
        
        # Agregar el texto 'DESENFOCADO' con el valor de la métrica centrado en el frame
        text1 = f'DESENFOCADO'
        text2 = f'Metrica: {focus_measure:.2f}'
        font = cv.FONT_HERSHEY_SIMPLEX
        text_size1 = cv.getTextSize(text1, font, 2, 3)[0]
        text_size2 = cv.getTextSize(text2, font, 1, 3)[0]
        text1_x = (frame_width // 2) - (text_size1[0] // 2)
        text1_y = (frame_height // 4) - text_size1[1] // 2
        text2_x = (frame_width // 2) - (text_size2[0] // 2)
        text2_y = (frame_height // 4) * 3 - text_size2[1] // 2
        cv.putText(frame, text1, (text1_x, text1_y), font, 2, (0, 0, 255), 3, cv.LINE_AA)
        cv.putText(frame, text2, (text2_x, text2_y), font, 1, (0, 0, 255), 3, cv.LINE_AA)
    else:
        # Dibujar un bounding box de color verde que ocupe todo el ancho y alto del frame
        cv.rectangle(frame, (start_x, start_y), (start_x + roi_w, start_y + roi_h), (0, 255, 0), 1)
        
        # Agregar el texto 'ENFOCADO' con el valor de la métrica centrado en el frame
        text1 = f'ENFOCADO'
        text2 = f'Metrica: {focus_measure:.2f}'
        font = cv.FONT_HERSHEY_SIMPLEX
        text_size1 = cv.getTextSize(text1, font, 2, 3)[0]
        text_size2 = cv.getTextSize(text2, font, 1, 3)[0]
        text1_x = (frame_width // 2) - (text_size1[0] // 2)
        text1_y = (frame_height // 4) - text_size1[1] // 2
        text2_x = (frame_width // 2) - (text_size2[0] // 2)
        text2_y = (frame_height // 4) * 3 - text_size2[1] // 2
        cv.putText(frame, text1, (text1_x, text1_y), font, 2, (0, 255, 0), 3, cv.LINE_AA)
        cv.putText(frame, text2, (text2_x, text2_y), font, 1, (0, 255, 0), 3, cv.LINE_AA)

    # Escribir el frame en el nuevo video
    out.write(frame)
    frame_count += 1

# Liberar los objetos VideoCapture y VideoWriter
cap.release()
out.release()
```

### Parte 1c: Medición sobre una Matriz de Enfoque (NxM)
Para esta parte vamos a dividir cada frame en una matriz de elementos rectangulares y a calcular la métrica para cada sección.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Definir el video y leer sus características
video_path = 'focus_video.mov'
cap = cv.VideoCapture(video_path)

# Obtener el ancho, alto y fps del video original
frame_width = int(cap.get(cv.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv.CAP_PROP_FPS)

# Valores para la matriz de enfoque
N_values = [3, 7]  # Probar varios valores para N y M
M_values = [3, 5]

# Definir el tamaño de la matriz que será menor que el tamaño total del frame
grid_width_ratio = 0.6  # Relación del ancho del área de la matriz con respecto al frame (60%)
grid_height_ratio = 0.6  # Relación del alto del área de la matriz con respecto al frame (60%)

# Definir los espaciados entre celdas
horizontal_padding = 10  # Espaciado entre celdas horizontalmente en píxeles
vertical_padding = 10  # Espaciado entre celdas verticalmente en píxeles

for N in N_values:
    for M in M_values:
        focus_scores_grid = []  # Lista para almacenar los valores de la métrica de enfoque de la matriz
        
        # Inicializar la matriz para almacenar las métricas de enfoque de cada celda a lo largo del video
        matriz_focos = np.empty((N, M), dtype=object)
        for i in range(N):
            for j in range(M):
                matriz_focos[i, j] = []

        # Crear VideoWriter para guardar el video con la matriz superpuesta
        output_path = f'focus_video_with_grid_matriz_{N}x{M}_LAPLACIANO.avi'
        fourcc = cv.VideoWriter_fourcc(*'XVID')
        out = cv.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

        # Reiniciar el video
        cap.set(cv.CAP_PROP_POS_FRAMES, 0)

        frame_count = 0  # Contador para el número de frames
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break

            # Convertir el frame a escala de grises
            gray_frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)

            # Definir la región del frame donde se colocará la matriz de enfoque
            grid_width = int(frame_width * grid_width_ratio)
            grid_height = int(frame_height * grid_height_ratio)

            start_x = (frame_width - grid_width) // 2
            start_y = (frame_height - grid_height) // 2

            # Ajustar el tamaño de cada celda para incluir el espaciado
            step_h = (grid_height - (N - 1) * vertical_padding) // N
            step_w = (grid_width - (M - 1) * horizontal_padding) // M

            total_focus = 0

            # Iterar sobre la matriz NxM para calcular la métrica de enfoque en cada celda
            for i in range(N):
                for j in range(M):
                    # Definir la ROI dentro del área de la matriz, teniendo en cuenta el padding
                    roi_x1 = start_x + j * (step_w + horizontal_padding)
                    roi_x2 = roi_x1 + step_w
                    roi_y1 = start_y + i * (step_h + vertical_padding)
                    roi_y2 = roi_y1 + step_h

                    roi = gray_frame[roi_y1:roi_y2, roi_x1:roi_x2]

                    # Calcular la varianza del Laplaciano en el frame
                    laplacian_var = cv.Laplacian(gray_frame, cv.CV_64F).var()
                    
                    matriz_focos[i, j].append(laplacian_var)

                    # Sumar las métricas para cada sección
                    total_focus += laplacian_var

            focus_scores_grid.append(total_focus)
            frame_count += 1

        # Graficar la evolución de la métrica sobre la matriz
        plt.plot(focus_scores_grid)
        plt.title(f'Evolución de la métrica de enfoque para la matriz {N}x{M}')
        plt.xlabel('Frame')
        plt.ylabel('Métrica de Enfoque Matriz')
        plt.show()

        # Reiniciar el video
        cap.set(cv.CAP_PROP_POS_FRAMES, 0)

        frame_count = 0
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break

            # Convertir el frame a escala de grises
            gray_frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)

            # Definir la región del frame donde se colocará la matriz de enfoque
            grid_width = int(frame_width * grid_width_ratio)
            grid_height = int(frame_height * grid_height_ratio)

            start_x = (frame_width - grid_width) // 2
            start_y = (frame_height - grid_height) // 2

            # Ajustar el tamaño de cada celda para incluir el espaciado
            step_h = (grid_height - (N - 1) * vertical_padding) // N
            step_w = (grid_width - (M - 1) * horizontal_padding) // M

            # Iterar sobre la matriz de métricas para mostrar en rojo o verde dependiendo del valor de enfoque
            for i in range(N):
                for j in range(M):
                    # Definir la ROI dentro del área de la matriz, teniendo en cuenta el padding
                    roi_x1 = start_x + j * (step_w + horizontal_padding)
                    roi_x2 = roi_x1 + step_w
                    roi_y1 = start_y + i * (step_h + vertical_padding)
                    roi_y2 = roi_y1 + step_h

                    # Determinar si el valor del enfoque es el máximo para esa celda
                    current_focus = matriz_focos[i, j][frame_count]
                    max_focus = max(matriz_focos[i, j])
                    if current_focus == max_focus:
                        color = (0, 255, 0)  # Verde para el valor máximo
                    else:
                        color = (0, 0, 255)  # Rojo para otros valores

                    # Dibujar el rectángulo sobre la ROI en el frame original
                    cv.rectangle(frame, (roi_x1, roi_y1), (roi_x2, roi_y2), color, 1)

                    # Colocar el valor de la métrica de enfoque dentro de la celda
                    text = f'{current_focus:.2f}'
                    text_size = cv.getTextSize(text, cv.FONT_HERSHEY_SIMPLEX, 0.3, 1)[0]
                    text_x = roi_x1 + (step_w - text_size[0]) // 2
                    text_y = roi_y1 + (step_h + text_size[1]) // 2
                    cv.putText(frame, text, (text_x, text_y), cv.FONT_HERSHEY_SIMPLEX, 0.3, color, 1, cv.LINE_AA)

            # Escribir el frame con la matriz superpuesta al nuevo video
            out.write(frame)

            frame_count += 1

        # Liberar el recurso `VideoWriter` después de completar cada video
        out.release()

# Liberar los recursos del video
cap.release()
cv.destroyAllWindows()
```
